```python
import sys
import numpy as np

import pygame
from pygame.locals import *
import math
import copy
pygame.init()
 
fps = 60
fpsClock = pygame.time.Clock()
 
width, height = 800, 800
scopeInc = 10
scope = 20
viewpos = [width/2-scope/2,height/2-scope/2]
grid = np.zeros(width*height).reshape(height,width)
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Conway's Game of Life")
BLACK = (0,0,0)
WHITE = (255,255,255)
#print(viewpos,scope)
pixels = np.array([])
class pixel():
    def __init__(self,x,y):
        self.x = x
        self.y = y
def draw_window():
    screen.fill(BLACK)
    for i in range(scope):
        inc = width/scope
        pygame.draw.line(screen,WHITE,(i*inc,0),(i*inc,height))
        pygame.draw.line(screen,WHITE,(0,i*inc),(width,i*inc))
        pygame.draw.line(screen,WHITE,(scope*inc-1,0),(scope*inc-1,height))
        pygame.draw.line(screen,WHITE,(0,scope*inc-1),(width,scope*inc-1))
    for pixel in pixels:
        if(pixel.x >= viewpos[0] and pixel.x < viewpos[0] + scope and pixel.y >= viewpos[1] and pixel.y < viewpos[1] + scope):
            x = pixel.x-viewpos[0]
            y = pixel.y-viewpos[1]
            inc = width/scope
            pygame.draw.rect(screen,WHITE,[x*inc,y*inc,inc,inc],width == 0)
            

def update():
    global grid,pixels
    kept = []
    new_grid = np.zeros(width*height).reshape(height,width)
    for p in pixels:
        x = p.x
        y = p.y
        miniGrid = grid[(y-1):(y+2),(x-1):(x+2)]
        alive = np.sum(miniGrid) - 1
        if(alive == 2 or alive == 3):
            new_grid[y][x] = 1
            kept.append([[y],[x]])
        for i in range(-1,2):
            for j in range(-1,2):
                miniGrid = grid[(y-1+i):(y+2+i),(x-1+j):(x+2+j)]
                alive = np.sum(miniGrid)
                if(alive == 3):
                    new_grid[(y+i)][(x+j)] = 1
                    kept.append([[y+i],[x+j]])
        

                    
                        
#     for remove in deadPixels:
#         pixels = np.delete(pixels,np.where(pixels==remove))
#     for add in revivedPixels:
#         pixels = np.append(pixels,add)
    #kept = np.unique(kept)
    yindex,xindex = np.where(new_grid == 1)
    pixels = np.array([])
    for i in range(len(yindex)):
        pixels = np.append(pixels,pixel(xindex[i],yindex[i]))
    grid = copy.deepcopy(new_grid)
    
        
# Game loop.
start_game = False
delta = 200
nextTime = 0
while True:
    global grid
    screen.fill((0, 0, 0))
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        elif event.type == pygame.KEYDOWN:
            if event.unicode == "=":
                scope -= scopeInc
                viewpos = [viewpos[0]+scopeInc/2,viewpos[1]+scopeInc/2]
            elif event.unicode == "-":
                scope += scopeInc
                viewpos = [viewpos[0]-scopeInc/2,viewpos[1]-scopeInc/2]
            elif event.key == pygame.K_LEFT:
                viewpos[0] -= scope/10
            elif event.key == pygame.K_RIGHT:
                viewpos[0] += scope/10
            elif event.key == pygame.K_UP:
                viewpos[1] -= scope/10
            elif event.key == pygame.K_DOWN:
                viewpos[1] += scope/10
            elif event.key == pygame.K_s:
                start_game = True
            elif event.key == pygame.K_r:
                scope = 20
                grid = np.zeros(width*height).reshape(height,width)
                viewpos = [width/2-scope/2,height/2-scope/2]
                start_game = False
                pixels = np.array([])
                
        elif event.type == pygame.MOUSEBUTTONDOWN and (start_game == False):
            mouse_presses = pygame.mouse.get_pressed()
            if mouse_presses[0]:
                pos = pygame.mouse.get_pos()
                relGridX = int(math.floor((pos[0]/width)*scope) + viewpos[0])
                relGridY = int(math.floor((pos[1]/height)*scope) + viewpos[1])
                flag = True
                for p in pixels:
                    if (p.x == relGridX and p.y == relGridY):
                        flag = False
                
                if (flag):
                    pixels = np.append(pixels,pixel(relGridX,relGridY))
                    grid[relGridY][relGridX] = 1
                
                
                
  
  # Update.
  
  # Draw.
    draw_window()
    if start_game:
        if pygame.time.get_ticks() >= nextTime:
            update()
            nextTime = pygame.time.get_ticks() + delta
    pygame.display.flip()
    fpsClock.tick(fps)
```
