```python
from tkinter import *
import time
import random
flag = False
inc = 100
count = 0

```

```python
class line:
    def __init__(self,p1,p2):
        self.p1 = p1
        self.p2 = p2
class point:
    def __init__(self,canvas,x,y,diameter):
        self.canvas = canvas
        self.x = x
        self.y = y
        self.r = diameter/2

window = Tk()
w = 1364
h = 768
canvas = Canvas(window,width=w,height=h)
canvas.pack()
dp = point(canvas,0,0,0)
colors = ["red", "orange", "blue", "violet","indigo","yellow"]
#points = [point(canvas,0,384,10),point(canvas,682,384,10),point(canvas,682,0,10),point(canvas,682,384,10),point(canvas,682,768,10),point(canvas,682,384,10),point(canvas,1364,384,10)]
lines = []
intermediates = []
path = []     
```

```python
def addPoint(event):
    points.append(point(canvas,event.x,event.y,10))
    if len(points) > 1:
            lines.append(line(points[-2],points[-1]))

def startAnimation(event):
    for i in range(len(points)-1):
        lines.append(line(points[i],points[i+1]))
    global flag
    flag = !(flag)
     
def reset(event):
    global dp
    global path
    global lines
    global points
    global intermediates
    global flag
    if(event.keycode == 82):
        flag = False
        path = []
        lines = []
        points = []
        intermediates = []
        
        dp = point(canvas,0,0,0)
        
    if(event.keycode == 84):
        flag = False
        lines = []
        points = []
        intermediates = []
        dp = point(canvas,0,0,0)
def reDraw():
    canvas.create_oval(dp.x-dp.r,dp.y-dp.r,dp.x+dp.r,dp.y+dp.r,fill = "red",width = 2)
    for i in points:
        canvas.create_oval(i.x-i.r,i.y-i.r,i.x+i.r,i.y+i.r,fill = "green")
    for lineArray in intermediates:
        for line in lineArray:
            canvas.create_line(line.p1.x,line.p1.y,line.p2.x,line.p2.y,fill = colors[lineArray.index(line)%len(colors)],width = 2)  
    for i in lines:
        canvas.create_line(i.p1.x,i.p1.y,i.p2.x,i.p2.y,fill = "green",width = 3)
    for i in range(len(path)-1):
        canvas.create_line(path[i].x,path[i].y,path[i+1].x,path[i+1].y,fill = "black",width = 2)



window.bind("<Button-1>",addPoint)
window.bind("<Button-3>",startAnimation)
window.bind("<KeyPress>",reset)




while True:
    global dp
    global path
    global lines
    global points
    global intermediates
    global flag
    window.update()
    canvas.delete('all')
    reDraw()
    if flag:
        if(count == 0):
            path = []
        intermediates = []
        intermediates.append(lines)
        while len(intermediates[-1]) > 1:
            newInt = []
            for i in range(len(intermediates[-1])-1):
                point1x = intermediates[-1][i].p1.x + (count/inc)*(intermediates[-1][i].p2.x-intermediates[-1][i].p1.x)
                point1y = intermediates[-1][i].p1.y + (count/inc)*(intermediates[-1][i].p2.y-intermediates[-1][i].p1.y)
            
                point2x = intermediates[-1][i+1].p1.x + (count/inc)*(intermediates[-1][i+1].p2.x-intermediates[-1][i+1].p1.x)
                point2y = intermediates[-1][i+1].p1.y + (count/inc)*(intermediates[-1][i+1].p2.y-intermediates[-1][i+1].p1.y)
                newInt.append(line(point(canvas,point1x,point1y,0),point(canvas,point2x,point2y,0)))
            intermediates.append(newInt)    
        pointx = intermediates[-1][0].p1.x + (count/inc)*(intermediates[-1][0].p2.x-intermediates[-1][0].p1.x)
        pointy = intermediates[-1][0].p1.y + (count/inc)*(intermediates[-1][0].p2.y-intermediates[-1][0].p1.y)
        dp.x = pointx
        dp.y = pointy
        dp.r = 5
        count += 1
        path.append(point(canvas,dp.x,dp.y,dp.r))
        if (count > inc):
            flag = False
            count = 0
        time.sleep(0.01)
    
    
    
        

    
window.mainloop()

```
